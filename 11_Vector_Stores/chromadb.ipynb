```python
from langchain_openai import OpenAIEmbeddings
from langchain_core.documents import Document
from langchain_chroma import Chroma
from dotenv import load_dotenv
```

```python
load_dotenv()
model=OpenAIEmbeddings()
```

#  Create sample Document objects with metadata

```python
doc1 = Document(
    page_content="Virat Kohli is one of the most successful and consistent batsmen in IPL history.",
    metadata={"team": "Royal Challengers Bangalore"}
)
doc2 = Document(
    page_content="Rohit Sharma has led Mumbai Indians to multiple IPL titles as an aggressive opener.",
    metadata={"team": "Mumbai Indians"}
)
doc3 = Document(
    page_content="MS Dhoni is renowned for his calm demeanor and brilliant finishing skills for Chennai Super Kings.",
    metadata={"team": "Chennai Super Kings"}
)
doc4 = Document(
    page_content="Jasprit Bumrah is a lethal death-over bowler playing crucial roles for Mumbai Indians.",
    metadata={"team": "Mumbai Indians"}
)
doc5 = Document(
    page_content="Ravindra Jadeja is a phenomenal all-rounder contributing with bat and ball for Chennai Super Kings.",
    metadata={"team": "Chennai Super Kings"}
)
```

#  Combine into a list

```python
docs = [doc1, doc2, doc3, doc4, doc5]
```

# Create and persist the Chroma database instance

```python
vector_store=Chroma(
    embedding_function=model,
    persist_directory="my_chroma_db",
    collection_name="sample"
)
```

# add document

```python
vector_store.add_documents(docs)
```

# view documents

```python
db_content=vector_store.get(include=["documents", "metadatas", "embeddings"])
db_content
```

```python
print("Stored Ids: ",db_content['ids'])
print("Stored Documents: ",db_content['documents'])
```

# search documents

```python
query="Who among these are a bowler?"
```

```python
result=vector_store.similarity_search(
    query=query,
    k=2
)
result
```

```python
for idx, doc in enumerate(result):
    print(f"Result {idx+1}: {doc.page_content}")
```

# search with similarty score

```python
results_with_score = vector_store.similarity_search_with_score(query=query, k=2)
results_with_score
```

```python
for doc, score in results_with_score:
    print(f"Score (Lower is better): {score} | Content: {doc.page_content}")
```

# metadata filtering

```python
filtered_results = vector_store.similarity_search(
    query="Find players",
    k=2,
    filter={"team": "Chennai Super Kings"}
)
```

```python
filtered_results
```

```python
for doc in filtered_results:
    print(f"Filtered Result: {doc.page_content} (Team: {doc.metadata['team']})")
```

# update document

```python
target_id = db_content["ids"][0]  # First element (e.g., Virat Kohli record) 'ids': ['8e057d9f-8803-4249-9672-57af24f67d68']
```

```python
# Define updated data structure
updated_doc1 = Document(
    page_content="Virat Kohli, the former captain of Royal Challengers Bangalore, is renowned for his aggressive leadership and consistency.",
    metadata={"team": "Royal Challengers Bangalore"}
)
```

```python
# Apply modification tracking
vector_store.update_document(document_id=target_id, document=updated_doc1)
```

```python
print("Stored Ids: ",db_content['ids'])
print("Stored Documents: ",db_content['documents'])
```

# delete documnet

```python
vector_store.delete(ids=[target_id])
```
